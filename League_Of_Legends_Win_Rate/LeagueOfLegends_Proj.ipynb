```python
import pandas as pd
import json
import os
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
from sklearn.preprocessing import MinMaxScaler
from sklearn import tree

import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns', 500)
```

#Modeling on Champion Names of each Match

```python
mcn = pd.read_csv('match_champ_names.csv')
mcn
```

```python
dummied = pd.get_dummies(mcn)
dummied
```

```python
X = dummied.drop(['Match ID', 'Team A Won?'], axis=1)
y = dummied['Team A Won?']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.2, random_state=42)
```

```python
mlp = MLPClassifier(random_state=42).fit(X_train, y_train)
mlp.score(X_val, y_val)
```

```python
clf = tree.DecisionTreeClassifier(random_state=42).fit(X_train, y_train)
clf.score(X_val, y_val)
```

```python
clf = RandomForestClassifier(max_depth = None, random_state=42).fit(X_train, y_train)
clf.score(X_val, y_val)
```

#Taking the best Dataframe and finding the best hyperparameters for each model

```python
X = champ_duo_bot_wr.drop(['Match ID', 'Team A Won?'], axis=1)
y = champ_duo_bot_wr['Team A Won?']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
param_dist = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 50)],
    'activation': ['logistic', 'tanh', 'relu'],
    'solver': ['lbfgs', 'sgd', 'adam'],
    'alpha': [1e-4, 1e-3, 1e-2, 1e-1],
    'learning_rate': ['constant', 'invscaling', 'adaptive'],
    'max_iter': randint(300, 500),
    'random_state': [42]
}

mlp = MLPClassifier()
mlp_cv = RandomizedSearchCV(mlp, param_distributions=param_dist, n_iter=10, cv=5, scoring='accuracy')
mlp_cv.fit(X_train, y_train)
```

```python
mlp_best_params = mlp_cv.best_params_
mlp_best_params
```

```python
print(mlp_cv.best_score_)
mlp_test_accuracy = mlp_cv.score(X_test, y_test)
```

```python
param_dist = {
    'n_estimators': randint(50, 200),
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 5),
    'bootstrap': [True, False],
    'random_state': [42]
}

rf = RandomForestClassifier()
rf_cv = RandomizedSearchCV(rf, param_distributions=param_dist, n_iter=10, cv=5, scoring='accuracy')
rf_cv.fit(X_train, y_train)
```

```python
rf_best_params = rf_cv.best_params_
rf_best_params
```

```python
print(rf_cv.best_score_)
rf_test_accuracy = rf_cv.score(X_test, y_test)
```

```python
param_dist = {
    'n_neighbors': [3, 5, 7, 10, 15],  # Example values for the number of neighbors
    'weights': ['uniform', 'distance'],  # Options for the weight function
    'metric': ['euclidean', 'manhattan', 'chebyshev']  # Distance metrics to consider
}

knn = KNeighborsClassifier()
knn_cv = RandomizedSearchCV(knn, param_distributions=param_dist, n_iter=10, cv=5, scoring='accuracy')
knn_cv.fit(X_train, y_train)
```

```python
knn_best_params = knn_cv.best_params_
knn_best_params
```

```python
print(knn_cv.best_score_)
knn_test_accuracy = knn_cv.score(X_test, y_test)
```
